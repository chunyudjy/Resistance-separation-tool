```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV
df = pd.read_csv("HPPC-protocol-LG.csv")

# 自动解析时间
df["Time Stamp"] = pd.to_datetime(
    df["Time Stamp"],
    format="mixed"
)

# 创建纵向子图
fig, axes = plt.subplots(
    3, 1,
    figsize=(14, 8),
    sharex=True
)

# Voltage
axes[0].plot(df["Time Stamp"], df["Voltage"])
axes[0].set_ylabel("Voltage")
axes[0].set_title("Voltage")

# Current
axes[1].plot(df["Time Stamp"], df["Current"])
axes[1].set_ylabel("Current")
axes[1].set_title("Current")

# Step
axes[2].step(
    df["Time Stamp"],
    df["Step"],
    where="post"
)
axes[2].set_ylabel("Step")
axes[2].set_title("Step")

# x轴标签
axes[2].set_xlabel("Time Stamp")

# 美化
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV
df = pd.read_csv("HPPC-protocol-LG.csv")

# 自动解析时间
df["Time Stamp"] = pd.to_datetime(
    df["Time Stamp"],
    format="mixed"
)

# 创建纵向子图
fig, axes = plt.subplots(
    5, 1,
    figsize=(14, 12),
    sharex=True
)

# -------------------------
# Voltage
# -------------------------
axes[0].plot(
    df["Time Stamp"],
    df["Voltage"]
)
axes[0].set_ylabel("Voltage")
axes[0].set_title("Voltage")

# -------------------------
# Current
# -------------------------
axes[1].plot(
    df["Time Stamp"],
    df["Current"]
)
axes[1].set_ylabel("Current")
axes[1].set_title("Current")

# -------------------------
# Step
# -------------------------
axes[2].step(
    df["Time Stamp"],
    df["Step"],
    where="post"
)
axes[2].set_ylabel("Step")
axes[2].set_title("Step")

# -------------------------
# Procedure
# -------------------------
procedure_vals = (
    df["Procedure"]
    .fillna("NA")
    .astype(str)
)

procedure_unique = procedure_vals.unique()

procedure_map = {
    v: i for i, v in enumerate(procedure_unique)
}

procedure_codes = procedure_vals.map(procedure_map)

axes[3].step(
    df["Time Stamp"],
    procedure_codes,
    where="post"
)

axes[3].set_yticks(range(len(procedure_unique)))
axes[3].set_yticklabels(procedure_unique)

axes[3].set_ylabel("Procedure")
axes[3].set_title("Procedure")

# -------------------------
# Status
# -------------------------
status_vals = (
    df["Status"]
    .fillna("NA")
    .astype(str)
)

status_unique = status_vals.unique()

status_map = {
    v: i for i, v in enumerate(status_unique)
}

status_codes = status_vals.map(status_map)

axes[4].step(
    df["Time Stamp"],
    status_codes,
    where="post"
)

axes[4].set_yticks(range(len(status_unique)))
axes[4].set_yticklabels(status_unique)

axes[4].set_ylabel("Status")
axes[4].set_title("Status")

# x轴标签
axes[4].set_xlabel("Time Stamp")

# 美化
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
